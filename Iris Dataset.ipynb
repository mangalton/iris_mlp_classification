```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
```

```python
iris_train = pd.read_csv(r"C:\Users\manga\Downloads\iris.data",names=["sepal-length","sepal-width","petal-length","petal-width","class"])
iris_train.head()
```

```python
iris_features = iris_train.copy()
iris_labels = iris_features.pop('class')
```

```python
iris_features = np.array(iris_features)
iris_labels = pd.get_dummies(iris_labels)
iris_labels
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test ,y_train,y_test = train_test_split(iris_features, iris_labels,test_size=0.2)
```

```python
model = tf.keras.Sequential([
    layers.Dense(64, activation ='relu',input_shape=(4,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(3, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
            loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(X_train, y_train,epochs=100,batch_size=32,verbose=1)

loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print("Test Loss:",loss)
print("Test Accuracy:", accuracy)
```

```python
predicted_indices = np.argmax(model.predict(X_test), axis=1)
predicted_labels = iris_labels.columns[predicted_indices]
# Print the predicted class labels
print("Predicted Labels:")
print(predicted_labels)

```

```python
true_indices = np.argmax(y_test, axis=1)
true_labels = iris_labels.columns[true_indices]
correct_predictions = (predicted_labels == true_labels)
correct_predictions
```
